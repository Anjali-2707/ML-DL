```python
import pandas as pd
```

```python
db=pd.read_csv("50_Startups.csv")
```

```python
db.info()
```

```python
db.columns
```

```python
X = db[['R&D Spend', 'Administration', 'Marketing Spend']]
```

```python
y = db["Profit"]
```

```python
state = db["State"]
```

```python
state_fi =pd.get_dummies(state ,drop_first=True)
```

```python
state_fi
```

```python
db_fi = pd.concat([state_fi,X],axis = 1)
```

```python
db_fi
```

```python
from sklearn.model_selection import train_test_split
```

```python
db_fi_train,db_fi_test,y_train,y_test = train_test_split(db_fi,y,test_size=0.20,random_state=42)
```

```python
db_fi_train.info
```

```python
db_fi_train=db_fi_train.values#to convert into pandas
```

```python
db_fi_train.reshape(-1,1)#to convert into 2D
```

```python
type(db_fi_train)
```

```python
db_fi_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(db_fi_train,y_train)
```

```python
model.coef_
```

```python
import joblib
```

```python
joblib.dump(model,"startups.pk1")
```

```python
db_fi_train.shape
```

```python
y_pred=model.predict(db_fi_test)
```
