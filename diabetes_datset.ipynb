```python
import pandas as pd
```

```python
db = pd.read_csv('pima-indians-diabetes.csv',header=None)
```

```python
y = db[8]
```

```python
X = db[[0,1,2,3,4,5,6,7]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
model.get_config()
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(units=10,
             bias_initializer="zeros",
             kernel_initializer="zeros",
             input_shape=(8, ),
             activation = "relu"))
```

```python
model.get_config()
```

```python
#8features * 10 neurons = 80 + 10 bias(10 neurons=10 bias)= 90 parameters
model.summary() 
```

```python
model.add(Dense(units=8,
             bias_initializer="zeros",
             kernel_initializer="zeros",
             activation = "relu"))
```

```python
model.get_config()
```

```python
model.add(Dense(units=8,
             bias_initializer="zeros",
             kernel_initializer="zeros",
             activation = "relu"))
```

```python
#input for 2nd layer is o/p of 1st layer 
#we have assign 8 neurons in 2nd layer
#10neurons(feature for 2nd layer)*8neurons(assigned in 2nd layer)=80
#80+8 bias(bias=neurons)= 88 parameters
model.summary()
```

```python
model.add(Dense(units=1,
             bias_initializer="zeros",
             kernel_initializer="zeros",
             activation = "sigmoid"))
```

```python
from keras.optimizers import Adam
```

```python
model.compile(optimizer=Adam(learning_rate=0.0000001),
             loss="binary_crossentropy",
             metrics=['accuracy'])
             
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
y_pred = model.predict(X_train)
```
